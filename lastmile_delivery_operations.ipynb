```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io

from google.colab import files
files.upload()
```

```python
datam=pd.read_csv('lastmile_delivery_operations_mit_lift_lab.csv')
datam
```

```python
#Eliminar columnas innecesarias
datam1=datam.drop(["_record_id","_updated_by","_title","_server_updated_at","_geometry","_latitude","_longitude","picture_of_the_parked_freight_vehicle","describe_the_accident","maximum_number_of_vehicles_in_the_traffic_jam"],axis=1)
```

```python
datam1.info()
```

```python
#Filtro para separar variables cualitativas
fdatam_cual=datam1.iloc[:,[0,1,2,3,4,6,7,8,9,10,11,12,15,16,17,18]]
```

```python
#Filtro para separar variables cuantitativas
fdatam_cuan=datam1.iloc[:,[5,13,14]]
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante forward fill" ("ffill")
fdatam_cual=fdatam_cual.fillna(method="bfill")
```

```python
#Identificar valores nulos por columna
valores_nulos=fdatam_cual.isnull().sum()
valores_nulos
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=fdatam_cuan

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
data3=fdatam_cuan[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3
```

```python
#Segundo método de sustitución de valores nulos
#Sustituir valores nulos con la media
data3=data3.fillna(round(fdatam_cuan.median(),1))
```

```python
#Identificar valores nulos por columna
valores_nulos=data3.isnull().sum()
valores_nulos
```
